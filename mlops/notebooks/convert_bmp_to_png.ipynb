```python
import os

import cv2
```

```python
def convert_file(filename):
    output_fname = filename.replace(".bmp", ".png")
    img = cv2.imread(filename)
    cv2.imwrite(output_fname, img)
    return output_fname
```

```python
def search_bmp(root_dir):
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for fname in filenames:
            if fname.endswith(".bmp"):
                res = convert_file(os.path.join(dirpath, fname))
                print("convert {} to {}".format(fname, res))
```

```python
search_bmp("/home/kitamura/dataset/NEUMetalSurfaceDefects")
```
